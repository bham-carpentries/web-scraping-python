```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
example_html = """
<!DOCTYPE html>
<html>
<head>
<title>Sample web page</title>
</head>
<body>
<h1>h1 Header #1</h1>
<p>This is a paragraph tag</p>
<h2>h2 Sub-header</h2>
<p>A new paragraph, now in the <b>sub-header</b></p>
<h1>h1 Header #2</h1>
<p>
This other paragraph has two hyperlinks,
one to <a href="https://carpentries.org/">The Carpentries homepage</a>,
and another to the
<a href="https://carpentries.org/workshops/past-workshops/">past workshops</a> page.
</p>
</body>
</html>
"""
```

```python
soup = BeautifulSoup(example_html, 'html.parser')
print(soup.prettify())
```

```python
print( soup.find('title') )
```

```python
print( soup.find('title').get_text())
```

```python
print( soup.find('h1').get_text() )
```

```python
links = soup.find_all('a')
print("Number of hyperlinks found: ", len(links))
print(links)                      
```

```python
for link in links:
    print(link.get('href'))
```

```python
first_link = {'element':str(links[0]),
              'url':links[0].get('href'),
              'text':links[0].get_text()}
```

```python
link_info_list = []
for item in links:
    dict_a = {}
    dict_a['element'] = str(item)
    dict_a['url'] = item.get('href')
    dict_a['text'] = item.get_text()
    link_info_list.append(dict_a)

links_df = pd.DataFrame(link_info_list, columns = ['element', 'url','text'])
print(links_df)
```

```python
links_df.to_csv('OutputCSV.csv',index=False)
```
